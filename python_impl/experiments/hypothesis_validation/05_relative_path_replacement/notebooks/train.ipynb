```python
from pathlib import Path
import json
import sys
ROOT = Path.cwd()
for parent in [ROOT] + list(ROOT.parents):
    if (parent / 'python_impl').exists() and (parent / 'README.md').exists():
        ROOT = parent
        break
if str(ROOT / 'python_impl') not in sys.path:
    sys.path.insert(0, str(ROOT / 'python_impl'))
print('ROOT =', ROOT)
from python_scripts.hypothesis_validation_common import SingleControlValidationConfig, build_single_control_dataset, evaluate_single_control_suite, resolve_stage_level, train_single_control_suite
STEP_DIR = ROOT / 'python_impl' / 'experiments' / 'hypothesis_validation' / '05_relative_path_replacement'
LEVEL = 'L1'
DEVICE = 'auto'
SEED = 20260401
FEATURE_KIND = 'relative'
stage_cfg = resolve_stage_level('05', LEVEL)
DATA_PATH = STEP_DIR / 'data' / f"single_control_relative_anechoic_{stage_cfg['level'].lower()}.h5"
RESULT_DIR = STEP_DIR / 'results' / stage_cfg['level'] / FEATURE_KIND
REFERENCE_SUMMARY = ROOT / 'python_impl' / 'experiments' / 'hypothesis_validation' / '03_single_control_hyperplane_anechoic' / 'results' / 'L1' / 'summary.json'
DATA_PATH
```

```python
cfg = SingleControlValidationConfig(num_samples=int(stage_cfg['num_samples']), seed=SEED, profile='anechoic')
summary = build_single_control_dataset(DATA_PATH, cfg)
summary
```

```python
train_summary = train_single_control_suite(
    DATA_PATH,
    RESULT_DIR,
    feature_kind=FEATURE_KIND,
    epochs=int(stage_cfg['epochs']),
    batch_size=128,
    device=DEVICE,
    seed=SEED,
    live_plot=True,
    stage_id='05',
    level=stage_cfg['level'],
)
train_summary
```

```python
eval_summary = evaluate_single_control_suite(DATA_PATH, RESULT_DIR, feature_kind=FEATURE_KIND, reference_summary_path=REFERENCE_SUMMARY, device=DEVICE, stage_id='05', level=stage_cfg['level'])
print(json.dumps(eval_summary, ensure_ascii=False, indent=2))
```
